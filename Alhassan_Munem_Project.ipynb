#Scatterplot with Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

np.random.seed(0)

# Convert tip_percentage to a 2D array
X = anime['score'].to_numpy().reshape(-1, 1)
y = anime['rank'].to_numpy()

# Fit the linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Extract coefficients
beta_0 = lin_reg.intercept_
beta_1 = lin_reg.coef_[0]
print(f"Intercept (β₀): {beta_0:.2f}")
print(f"Slope (β₁): {beta_1:.2f}")

# Create the plot
plt.figure(figsize=(8, 5))
sns.regplot(x=X[:, 0], y=y, scatter=True, ci=None, line_kws={"color": "red"})

# Create the equation text and annotate the plot
equation_text = f"$y = {beta_0:.2f} + {beta_1:.2f}x$"
plt.text(0.05, 0.95, equation_text, transform=plt.gca().transAxes, fontsize=12, color="red", verticalalignment='top')

plt.title('Linear Regression Fit')
plt.xlabel('Score')
plt.ylabel('Rank')
plt.show()
```

```python
def ols_equations(value):
    y_new = beta_0 + beta_1 * value
    return y_new.round(2)

x_val = 7.5
print("The Predicted rank using a score of 7.5 =", ols_equations(x_val))
```

```python
corr = np.corrcoef(X.flatten(), y)[0, 1]

print(f"Correlation Coefficient (r): {corr:.2f}")
```

```python
abs(corr) #strong correlation if right
```

The equation of our linear model is y = 36814.95 - 4557.75x. Using the equation and x = 7.5, the predicted rank is found to be 2631.83. The Correlation Coefficient (r): -0.98, therefore |r| = 0.98 which shows  strong correlation.

**2.   the probability of anime being ranked less than 50 given they are a movie**

```python
contingency_table = pd.crosstab(anime['type'], anime['rank'] < 50)
contingency_table
```

```python
conditional_probabilities = contingency_table.div(contingency_table.sum(axis=1), axis=0)
print(conditional_probabilities)
```

The probability of anime being ranked less than 50 given they are a movie = 0.005
